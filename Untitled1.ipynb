```python
from google.colab import files
import io, numpy as np, pandas as pd, matplotlib.pyplot as plt
from scipy.optimize import minimize

uploaded = files.upload()
filename = list(uploaded.keys())[0]
```

```python
data = pd.read_csv(io.BytesIO(uploaded[filename]))
data.columns = [c.strip().lower() for c in data.columns]
print("Columns found:", data.columns.tolist())
print(data.head())

x_col = [c for c in data.columns if 'x' in c][0]
y_col = [c for c in data.columns if 'y' in c][0]
```

```python
if 't' in data.columns:
    t = data['t'].values
else:
    t = np.linspace(6, 60, len(data))

X_data = data[x_col].values
Y_data = data[y_col].values
```

```python
def model(t, theta, M, X0):
    X = t * np.cos(theta) - np.exp(M * np.abs(t)) * np.sin(0.3*t) * np.sin(theta) + X0
    Y = 42 + t * np.sin(theta) + np.exp(M * np.abs(t)) * np.sin(0.3*t) * np.cos(theta)
    return X, Y
```

```python
def error(params):
    theta, M, X0 = params
    X_pred, Y_pred = model(t, theta, M, X0)
    return np.sum(np.abs(X_pred - X_data) + np.abs(Y_pred - Y_data))
```

```python
initial_guess = [np.radians(10), 0.0, 10.0]
bounds = [(0, np.radians(50)), (-0.05, 0.05), (0, 100)]
result = minimize(error, initial_guess, bounds=bounds)

theta_best, M_best, X0_best = result.x
print("\n✅ Optimization complete!")
print(f"Theta (degrees): {np.degrees(theta_best):.4f}")
print(f"M: {M_best:.6f}")
print(f"X0: {X0_best:.4f}")
```

```python
X_pred, Y_pred = model(t, theta_best, M_best, X0_best)
plt.figure(figsize=(8,6))
plt.plot(X_data, Y_data, 'bo', label='Data Points')
plt.plot(X_pred, Y_pred, 'r-', linewidth=2, label='Fitted Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Parametric Curve Fit')
plt.legend()
plt.grid(True)
plt.show()
```
